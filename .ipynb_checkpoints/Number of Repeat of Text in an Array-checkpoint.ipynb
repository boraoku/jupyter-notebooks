```python
import matplotlib.pyplot as plt
from collections import Counter

def create_array_from_multiline_text(text):
    # Split the text into lines
    lines = text.split("\n")

    # Remove any empty lines
    lines = [line for line in lines if line.strip()]

    # Return the resulting array
    return lines

def plot_string_frequency(arr, title):
    # Count the frequency of each string in the array
    freq = Counter(arr)

    # Get the unique strings and their frequencies
    unique_strings = list(freq.keys())
    frequencies = list(freq.values())

    # Create a bar chart
    plt.bar(unique_strings, frequencies)

    for i, v in enumerate(frequencies):
        plt.text(i, v + 0.1, str(v), ha='center')
    
    # Set the title and axis labels
    plt.title(title)
    plt.xlabel("Ground Conditions")
    plt.ylabel("Frequency")
    
    plt.xticks(rotation=90)

    # Show the plot
    plt.show()
```
